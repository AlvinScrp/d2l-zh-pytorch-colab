è™½ç„¶ç°ä»£çš„æ·±åº¦å­¦ä¹ æ¡†æ¶å‡ ä¹å¯ä»¥è‡ªåŠ¨åŒ–åœ°è¿›è¡Œæ‰€æœ‰è¿™äº›å·¥ä½œï¼Œä½†ä»é›¶å¼€å§‹å®ç°å¯ä»¥ç¡®ä¿æˆ‘ä»¬çœŸæ­£çŸ¥é“è‡ªå·±åœ¨åšä»€ä¹ˆã€‚

## ç”Ÿæˆæ•°æ®é›†

ä¸ºäº†ç®€å•èµ·è§ï¼Œæˆ‘ä»¬å°†[**æ ¹æ®å¸¦æœ‰å™ªå£°çš„çº¿æ€§æ¨¡å‹æ„é€ ä¸€ä¸ªäººé€ æ•°æ®é›†ã€‚**]
æˆ‘ä»¬çš„ä»»åŠ¡æ˜¯ä½¿ç”¨è¿™ä¸ªæœ‰é™æ ·æœ¬çš„æ•°æ®é›†æ¥æ¢å¤è¿™ä¸ªæ¨¡å‹çš„å‚æ•°ã€‚
æˆ‘ä»¬å°†ä½¿ç”¨ä½ç»´æ•°æ®ï¼Œè¿™æ ·å¯ä»¥å¾ˆå®¹æ˜“åœ°å°†å…¶å¯è§†åŒ–ã€‚
åœ¨ä¸‹é¢çš„ä»£ç ä¸­ï¼Œæˆ‘ä»¬ç”Ÿæˆä¸€ä¸ªåŒ…å«1000ä¸ªæ ·æœ¬çš„æ•°æ®é›†ï¼Œ
æ¯ä¸ªæ ·æœ¬åŒ…å«ä»æ ‡å‡†æ­£æ€åˆ†å¸ƒä¸­é‡‡æ ·çš„2ä¸ªç‰¹å¾ã€‚
æˆ‘ä»¬çš„åˆæˆæ•°æ®é›†æ˜¯ä¸€ä¸ªçŸ©é˜µ$\mathbf{X}\in \mathbb{R}^{1000 \times 2}$ã€‚

(**æˆ‘ä»¬ä½¿ç”¨çº¿æ€§æ¨¡å‹å‚æ•°$\mathbf{w} = [2, -3.4]^\top$ã€$b = 4.2$
å’Œå™ªå£°é¡¹$\epsilon$ç”Ÿæˆæ•°æ®é›†åŠå…¶æ ‡ç­¾ï¼š

$$\mathbf{y}= \mathbf{X} \mathbf{w} + b + \mathbf\epsilon.$$
**)

$\epsilon$å¯ä»¥è§†ä¸ºæ¨¡å‹é¢„æµ‹å’Œæ ‡ç­¾æ—¶çš„æ½œåœ¨è§‚æµ‹è¯¯å·®ã€‚
åœ¨è¿™é‡Œæˆ‘ä»¬è®¤ä¸ºæ ‡å‡†å‡è®¾æˆç«‹ï¼Œå³$\epsilon$æœä»å‡å€¼ä¸º0çš„æ­£æ€åˆ†å¸ƒã€‚
ä¸ºäº†ç®€åŒ–é—®é¢˜ï¼Œæˆ‘ä»¬å°†æ ‡å‡†å·®è®¾ä¸º0.01ã€‚
ä¸‹é¢çš„ä»£ç ç”Ÿæˆåˆæˆæ•°æ®é›†ã€‚


```python
def synthetic_data(w, b, num_examples):
    """ç”Ÿæˆy=Xw+b+å™ªå£°"""
    X = torch.normal(0, 1, (num_examples, len(w)))
    y = torch.matmul(X, w) + b
    y += torch.normal(0, 0.01, y.shape)
    return X, y.reshape((-1, 1))
```

## è¯»å–æ•°æ®é›†

å›æƒ³ä¸€ä¸‹ï¼Œè®­ç»ƒæ¨¡å‹æ—¶è¦å¯¹æ•°æ®é›†è¿›è¡Œéå†ï¼Œæ¯æ¬¡æŠ½å–ä¸€å°æ‰¹é‡æ ·æœ¬ï¼Œå¹¶ä½¿ç”¨å®ƒä»¬æ¥æ›´æ–°æˆ‘ä»¬çš„æ¨¡å‹ã€‚
ç”±äºè¿™ä¸ªè¿‡ç¨‹æ˜¯è®­ç»ƒæœºå™¨å­¦ä¹ ç®—æ³•çš„åŸºç¡€ï¼Œæ‰€ä»¥æœ‰å¿…è¦å®šä¹‰ä¸€ä¸ªå‡½æ•°ï¼Œ
è¯¥å‡½æ•°èƒ½æ‰“ä¹±æ•°æ®é›†ä¸­çš„æ ·æœ¬å¹¶ä»¥å°æ‰¹é‡æ–¹å¼è·å–æ•°æ®ã€‚

åœ¨ä¸‹é¢çš„ä»£ç ä¸­ï¼Œæˆ‘ä»¬[**å®šä¹‰ä¸€ä¸ª`data_iter`å‡½æ•°ï¼Œ
è¯¥å‡½æ•°æ¥æ”¶æ‰¹é‡å¤§å°ã€ç‰¹å¾çŸ©é˜µå’Œæ ‡ç­¾å‘é‡ä½œä¸ºè¾“å…¥ï¼Œç”Ÿæˆå¤§å°ä¸º`batch_size`çš„å°æ‰¹é‡**]ã€‚
æ¯ä¸ªå°æ‰¹é‡åŒ…å«ä¸€ç»„ç‰¹å¾å’Œæ ‡ç­¾ã€‚


```python
def data_iter(batch_size, features, labels):
    num_examples = len(features)
    indices = list(range(num_examples))
    # è¿™äº›æ ·æœ¬æ˜¯éšæœºè¯»å–çš„ï¼Œæ²¡æœ‰ç‰¹å®šçš„é¡ºåº
    random.shuffle(indices) # æ‰“ä¹±é¡ºåº
    for i in range(0, num_examples, batch_size):
        batch_indices = torch.tensor(
            indices[i: min(i + batch_size, num_examples)])
        print(batch_indices)
        yield features[batch_indices], labels[batch_indices]
```

é€šå¸¸ï¼Œæˆ‘ä»¬åˆ©ç”¨GPUå¹¶è¡Œè¿ç®—çš„ä¼˜åŠ¿ï¼Œå¤„ç†åˆç†å¤§å°çš„â€œå°æ‰¹é‡â€ã€‚
æ¯ä¸ªæ ·æœ¬éƒ½å¯ä»¥å¹¶è¡Œåœ°è¿›è¡Œæ¨¡å‹è®¡ç®—ï¼Œä¸”æ¯ä¸ªæ ·æœ¬æŸå¤±å‡½æ•°çš„æ¢¯åº¦ä¹Ÿå¯ä»¥è¢«å¹¶è¡Œè®¡ç®—ã€‚
GPUå¯ä»¥åœ¨å¤„ç†å‡ ç™¾ä¸ªæ ·æœ¬æ—¶ï¼Œæ‰€èŠ±è´¹çš„æ—¶é—´ä¸æ¯”å¤„ç†ä¸€ä¸ªæ ·æœ¬æ—¶å¤šå¤ªå¤šã€‚

æˆ‘ä»¬ç›´è§‚æ„Ÿå—ä¸€ä¸‹å°æ‰¹é‡è¿ç®—ï¼šè¯»å–ç¬¬ä¸€ä¸ªå°æ‰¹é‡æ•°æ®æ ·æœ¬å¹¶æ‰“å°ã€‚
æ¯ä¸ªæ‰¹é‡çš„ç‰¹å¾ç»´åº¦æ˜¾ç¤ºæ‰¹é‡å¤§å°å’Œè¾“å…¥ç‰¹å¾æ•°ã€‚
åŒæ ·çš„ï¼Œæ‰¹é‡çš„æ ‡ç­¾å½¢çŠ¶ä¸`batch_size`ç›¸ç­‰ã€‚


```python
batch_size = 10

for X, y in data_iter(batch_size, features, labels):
    print(X, '\n', y)
    break
```

ä¸Šé¢å®ç°çš„è¿­ä»£å¯¹æ•™å­¦æ¥è¯´å¾ˆå¥½ï¼Œä½†å®ƒçš„æ‰§è¡Œæ•ˆç‡å¾ˆä½ï¼Œå¯èƒ½ä¼šåœ¨å®é™…é—®é¢˜ä¸Šé™·å…¥éº»çƒ¦ã€‚
ä¾‹å¦‚ï¼Œå®ƒè¦æ±‚æˆ‘ä»¬å°†æ‰€æœ‰æ•°æ®åŠ è½½åˆ°å†…å­˜ä¸­ï¼Œå¹¶æ‰§è¡Œå¤§é‡çš„éšæœºå†…å­˜è®¿é—®ã€‚
åœ¨æ·±åº¦å­¦ä¹ æ¡†æ¶ä¸­å®ç°çš„å†…ç½®è¿­ä»£å™¨æ•ˆç‡è¦é«˜å¾—å¤šï¼Œ

## åˆå§‹åŒ–æ¨¡å‹å‚æ•°

[**åœ¨æˆ‘ä»¬å¼€å§‹ç”¨å°æ‰¹é‡éšæœºæ¢¯åº¦ä¸‹é™ä¼˜åŒ–æˆ‘ä»¬çš„æ¨¡å‹å‚æ•°ä¹‹å‰**]ï¼Œ

åœ¨ä¸‹é¢çš„ä»£ç ä¸­ï¼Œæˆ‘ä»¬é€šè¿‡ä»å‡å€¼ä¸º0ã€æ ‡å‡†å·®ä¸º0.01çš„æ­£æ€åˆ†å¸ƒä¸­é‡‡æ ·éšæœºæ•°æ¥åˆå§‹åŒ–æƒé‡ï¼Œ
å¹¶å°†åç½®åˆå§‹åŒ–ä¸º0ã€‚


```python
w = torch.normal(0, 0.01, size=(2,1), requires_grad=True)
b = torch.zeros(1, requires_grad=True)
w,b
```

åœ¨åˆå§‹åŒ–å‚æ•°ä¹‹åï¼Œæˆ‘ä»¬çš„ä»»åŠ¡æ˜¯æ›´æ–°è¿™äº›å‚æ•°ï¼Œç›´åˆ°è¿™äº›å‚æ•°è¶³å¤Ÿæ‹Ÿåˆæˆ‘ä»¬çš„æ•°æ®ã€‚
æ¯æ¬¡æ›´æ–°éƒ½éœ€è¦è®¡ç®—æŸå¤±å‡½æ•°å…³äºæ¨¡å‹å‚æ•°çš„æ¢¯åº¦ã€‚
æœ‰äº†è¿™ä¸ªæ¢¯åº¦ï¼Œæˆ‘ä»¬å°±å¯ä»¥å‘å‡å°æŸå¤±çš„æ–¹å‘æ›´æ–°æ¯ä¸ªå‚æ•°ã€‚
å› ä¸ºæ‰‹åŠ¨è®¡ç®—æ¢¯åº¦å¾ˆæ¯ç‡¥è€Œä¸”å®¹æ˜“å‡ºé”™ï¼Œæ‰€ä»¥æ²¡æœ‰äººä¼šæ‰‹åŠ¨è®¡ç®—æ¢¯åº¦ã€‚
æˆ‘ä»¬ä½¿ç”¨ :numref:`sec_autograd`ä¸­å¼•å…¥çš„è‡ªåŠ¨å¾®åˆ†æ¥è®¡ç®—æ¢¯åº¦ã€‚

## å®šä¹‰æ¨¡å‹

æ¥ä¸‹æ¥ï¼Œæˆ‘ä»¬å¿…é¡»[**å®šä¹‰æ¨¡å‹ï¼Œå°†æ¨¡å‹çš„è¾“å…¥å’Œå‚æ•°åŒæ¨¡å‹çš„è¾“å‡ºå…³è”èµ·æ¥ã€‚**]
å›æƒ³ä¸€ä¸‹ï¼Œè¦è®¡ç®—çº¿æ€§æ¨¡å‹çš„è¾“å‡ºï¼Œ
æˆ‘ä»¬åªéœ€è®¡ç®—è¾“å…¥ç‰¹å¾$\mathbf{X}$å’Œæ¨¡å‹æƒé‡$\mathbf{w}$çš„çŸ©é˜µ-å‘é‡ä¹˜æ³•ååŠ ä¸Šåç½®$b$ã€‚
æ³¨æ„ï¼Œä¸Šé¢çš„$\mathbf{Xw}$æ˜¯ä¸€ä¸ªå‘é‡ï¼Œè€Œ$b$æ˜¯ä¸€ä¸ªæ ‡é‡ã€‚
å›æƒ³ä¸€ä¸‹ :numref:`subsec_broadcasting`ä¸­æè¿°çš„å¹¿æ’­æœºåˆ¶ï¼š
å½“æˆ‘ä»¬ç”¨ä¸€ä¸ªå‘é‡åŠ ä¸€ä¸ªæ ‡é‡æ—¶ï¼Œæ ‡é‡ä¼šè¢«åŠ åˆ°å‘é‡çš„æ¯ä¸ªåˆ†é‡ä¸Šã€‚


```python
def linreg(X, w, b):
    """çº¿æ€§å›å½’æ¨¡å‹"""
    return torch.matmul(X, w) + b
```

## [**å®šä¹‰æŸå¤±å‡½æ•°**]

å› ä¸ºéœ€è¦è®¡ç®—æŸå¤±å‡½æ•°çš„æ¢¯åº¦ï¼Œæ‰€ä»¥æˆ‘ä»¬åº”è¯¥å…ˆå®šä¹‰æŸå¤±å‡½æ•°ã€‚
è¿™é‡Œæˆ‘ä»¬ä½¿ç”¨ :numref:`sec_linear_regression`ä¸­æè¿°çš„å¹³æ–¹æŸå¤±å‡½æ•°ã€‚
åœ¨å®ç°ä¸­ï¼Œæˆ‘ä»¬éœ€è¦å°†çœŸå®å€¼`y`çš„å½¢çŠ¶è½¬æ¢ä¸ºå’Œé¢„æµ‹å€¼`y_hat`çš„å½¢çŠ¶ç›¸åŒã€‚


```python
def squared_loss(y_hat, y):
    """å‡æ–¹æŸå¤±"""
    return (y_hat - y.reshape(y_hat.shape)) ** 2 / 2
```

## (**å®šä¹‰ä¼˜åŒ–ç®—æ³•**)

æ­£å¦‚æˆ‘ä»¬åœ¨ :numref:`sec_linear_regression`ä¸­è®¨è®ºçš„ï¼Œçº¿æ€§å›å½’æœ‰è§£æè§£ã€‚
å°½ç®¡çº¿æ€§å›å½’æœ‰è§£æè§£ï¼Œä½†æœ¬ä¹¦ä¸­çš„å…¶ä»–æ¨¡å‹å´æ²¡æœ‰ã€‚
è¿™é‡Œæˆ‘ä»¬ä»‹ç»å°æ‰¹é‡éšæœºæ¢¯åº¦ä¸‹é™ã€‚

åœ¨æ¯ä¸€æ­¥ä¸­ï¼Œä½¿ç”¨ä»æ•°æ®é›†ä¸­éšæœºæŠ½å–çš„ä¸€ä¸ªå°æ‰¹é‡ï¼Œç„¶åæ ¹æ®å‚æ•°è®¡ç®—æŸå¤±çš„æ¢¯åº¦ã€‚
æ¥ä¸‹æ¥ï¼Œæœç€å‡å°‘æŸå¤±çš„æ–¹å‘æ›´æ–°æˆ‘ä»¬çš„å‚æ•°ã€‚
ä¸‹é¢çš„å‡½æ•°å®ç°å°æ‰¹é‡éšæœºæ¢¯åº¦ä¸‹é™æ›´æ–°ã€‚
è¯¥å‡½æ•°æ¥å—æ¨¡å‹å‚æ•°é›†åˆã€å­¦ä¹ é€Ÿç‡å’Œæ‰¹é‡å¤§å°ä½œä¸ºè¾“å…¥ã€‚æ¯
ä¸€æ­¥æ›´æ–°çš„å¤§å°ç”±å­¦ä¹ é€Ÿç‡`lr`å†³å®šã€‚
å› ä¸ºæˆ‘ä»¬è®¡ç®—çš„æŸå¤±æ˜¯ä¸€ä¸ªæ‰¹é‡æ ·æœ¬çš„æ€»å’Œï¼Œæ‰€ä»¥æˆ‘ä»¬ç”¨æ‰¹é‡å¤§å°ï¼ˆ`batch_size`ï¼‰
æ¥è§„èŒƒåŒ–æ­¥é•¿ï¼Œè¿™æ ·æ­¥é•¿å¤§å°å°±ä¸ä¼šå–å†³äºæˆ‘ä»¬å¯¹æ‰¹é‡å¤§å°çš„é€‰æ‹©ã€‚


```python
def sgd(params, lr, batch_size):
    """å°æ‰¹é‡éšæœºæ¢¯åº¦ä¸‹é™"""
    with torch.no_grad():
        for param in params:
            param -= lr * param.grad / batch_size
            param.grad.zero_()
```

## è®­ç»ƒ

ç°åœ¨æˆ‘ä»¬å·²ç»å‡†å¤‡å¥½äº†æ¨¡å‹è®­ç»ƒæ‰€æœ‰éœ€è¦çš„è¦ç´ ï¼Œå¯ä»¥å®ç°ä¸»è¦çš„[**è®­ç»ƒè¿‡ç¨‹**]éƒ¨åˆ†äº†ã€‚
